```python
# Import libraries needed
import numpy as np
import random as rd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate input and a slightly random output
Xinput = np.arange(100)
Yinput = np.multiply(Xinput, 6 + 6 * np.random.rand(Xinput.size))

print(Xinput)
print(Yinput)
```

```python
# Plot the graph of input and output
plt.plot(Xinput, Yinput)
plt.show()
```

```python
mean = sum(Yinput) / Yinput.size
print ('The mean values is:', mean)

# Initialize the weights / bias randomly. In this case, I wanted to test for
# extreme cases, hence the exponential guess
B = np.arange(Xinput.size) + 0.1 * np.random.rand(Xinput.size)
Ypredict = Xinput * B

plt.plot(Xinput, Ypredict, 'ro', label = 'Initial prediction')
plt.plot(Xinput, Yinput, 'g+',  label = 'Expected output')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Initial prediction')
plt.show()
```

```python
# Function to calculate the average root mean squared error of the prediction
# What this function does is to take the actualy output and subtract it by the prediction and square it.
# Then the average error is returned.
def costFunction(Xinput, Yinput, B):
    cost = 0
    for i in (Xinput):
        cost += (Yinput[i] - Xinput[i] * B[i]) **  2
    
    totalCost = cost / Xinput.size
    return totalCost
```

```python
# Function to update the weights of the prediction.
def updateWeight(Xinput, Yinput, B, learningRate):
    weightDeriv = 0
    
    for i in (Xinput):
        # Derivative of the cost function with respect to B
        weightDeriv = -2 * Xinput[i] * (Yinput[i] - (Xinput[i] * B[i])) 
        
        #Update the new values of the weights
        # Negative in the front because we need to lessen the value of B when the slope is >0, and vice versa.
        B[i] = float(B[i]) - (weightDeriv / Xinput.size) * learningRate
    
    return B
```

```python
# Run linear regression with 30 iterations with learning rate of 0.005
# Initialize an empty array costChange to keep track of our cost per iteration
costChange = []
Iteration = 0

for i in range (1,30):
    B = updateWeight(Xinput, Yinput, B, 0.005);
    cost = costFunction(Xinput, Yinput, B);
    costChange.append(cost)
    Iteration += 1

plt.plot(np.arange(Iteration), costChange)
plt.xlabel('Number of iterations')
plt.ylabel('Cost Function')
plt.title('Cost vs Number of Iteartions')
plt.show()
```
