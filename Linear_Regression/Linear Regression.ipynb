```python
import numpy as np
import random as rd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Xinput = np.arange(100)
Yinput = np.multiply(Xinput, 6 + 6 * np.random.rand(Xinput.size))

print(Xinput)
print(Yinput)
```

```python
plt.plot(Xinput, Yinput)
plt.show()
```

```python
mean = sum(Yinput) / Yinput.size
print ('The mean values is:', mean)

B = np.arange(Xinput.size) + 0.1 * np.random.rand(Xinput.size)
Ypredict = Xinput * B

plt.plot(Xinput, Ypredict, 'ro', label = 'Initial prediction')
plt.plot(Xinput, Yinput, 'g+',  label = 'Expected output')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Initial prediction')
plt.show()
```

```python
def costFunction(Xinput, Yinput, B):
    cost = 0
    for i in (Xinput):
        cost += (Yinput[i] - Xinput[i] * B[i]) **  2
    
    totalCost = cost / Xinput.size
    return totalCost
```

```python
def updateWeight(Xinput, Yinput, B, learningRate):
    weightDeriv = 0
    
    for i in (Xinput):
        weightDeriv = -2 * Xinput[i] * (Yinput[i] - (Xinput[i] * B[i]))
        B[i] = float(B[i]) - (weightDeriv / Xinput.size) * learningRate
    
    return B
```

```python
costChange = []
I = 0

for i in range (1,30):
    B = updateWeight(Xinput, Yinput, B, 0.005);
    cost = costFunction(Xinput, Yinput, B);
    costChange.append(cost)
    I += 1

plt.plot(np.arange(I), costChange)
plt.xlabel('Number of iterations')
plt.ylabel('Cost Function')
plt.title('Cost vs Number of Iteartions')
plt.show()
```
